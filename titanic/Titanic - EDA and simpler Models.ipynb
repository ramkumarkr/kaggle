```python
import pandas as pd
import matplotlib.pyplot as plt

train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
combined_data = train_data.append(test_data,sort=False)
```

```python
train_data['Cabin'] = train_data['Cabin'].fillna('DummyCabin')
test_data['Cabin'] = test_data['Cabin'].fillna('DummyCabin')
```

```python
combined_data['Age'].mean()
```

```python
train_data['Age'] = train_data['Age'].fillna(combined_data['Age'].mean())
test_data['Age'] = test_data['Age'].fillna(combined_data['Age'].mean())
```

```python
test_data[test_data['Fare'].isna()]
```

```python
train_data['Fare'] = train_data['Fare'].fillna(0)
test_data['Fare'] = test_data['Fare'].fillna(0)
```

```python
train_data[train_data['Embarked'].isna()]
```

```python
combined_data['Embarked'].unique()
```

```python
plt.hist(combined_data['Embarked'])
```

```python
train_data['Embarked'] = train_data['Fare'].fillna(0)
test_data['Embarked'] = test_data['Fare'].fillna(0)
```

```python
train_data.columns
```

```python
train_data = train_data.dropna(axis=0)
```

```python
train_data.describe()
```

```python
Y = train_data.Survived
```

```python
train_data_features = ['Pclass','Age','SibSp','Parch','Fare']
X = train_data[train_data_features]
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeRegressor 
titanic_dtree_model = DecisionTreeRegressor(random_state=1)

titanic_dtree_model.fit(X,Y)

train_predicted = titanic_dtree_model.predict(X)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(Y, train_predicted)
```
